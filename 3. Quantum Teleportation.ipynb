# Quantum Teleportation

You can't copy data from 1 qubit to another with conventional way because measuring it will collapse the qubit. Thus we have to teleport 

```python
import qiskit as q
from qiskit import IBMQ
from qiskit.tools.monitor import job_monitor
from qiskit.tools.visualization import plot_bloch_multivector, plot_histogram
```

```python
circuit = q.QuantumCircuit(3, 3)
```

```python
circuit.draw(output = 'mpl')
```

```python
circuit.x(0)
circuit.barrier()
circuit.draw(output = 'mpl')
```

```python
circuit.h(1)
circuit.cx(1, 2)
```

```python
circuit.draw(output = 'mpl')
```

```python
circuit.cx(0, 1)
circuit.h(0)
circuit.draw(output = 'mpl')
```

```python
circuit.barrier()
circuit.measure([0, 1], [0, 1])
circuit.draw(output = 'mpl')
```

```python
circuit.barrier()
circuit.cx(1, 2)
circuit.cz(0, 2)
circuit.draw(output = 'mpl')
```

```python
circuit.measure(2, 2)
simulator = q.Aer.get_backend('qasm_simulator')
results = q.execute(circuit, backend = simulator, shots = 1000).result()
counts = results.get_counts()
plot_histogram(counts)
```
